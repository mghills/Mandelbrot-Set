```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")
```

```python
x = [x / 1000.0 for x in range(-2000, 2001, 1)]
y = [y / 1000.0 for y in range(-2000, 2001, 1)]
```

```python
def inSet(c):
    z = 0
    for i in range(1, 100):
        z = pow(z, 2) + c
        if abs(z) > 2:
            return False
        
    return True
```

```python
def calcC(x, y):
    return complex(x, y)
```

```python
def getSet(x_list, y_list):
    x_set = list()
    y_set = list()
    for x in x_list:
        for y in y_list:
            if inSet(calcC(x, y)):
                x_set.append(x)
                y_set.append(y)

    return x_set, y_set
```

```python
x_set, y_set = getSet(x, y)
```

```python
# Plot the time series data
plt.figure(figsize=(20, 20))
plt.scatter(x_set, y_set, s=0.2)
plt.title('Mandelbrot Set')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
